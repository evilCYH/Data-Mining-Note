```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv('./data/iris.csv')
iris
```

```python
y1 = iris[iris['Species'] == 'setosa']['Sepal.Length']
y2 = iris[iris['Species'] == 'versicolor']['Sepal.Length']
y3 = iris[iris['Species'] == 'virginica']['Sepal.Length']

x = np.linspace(1,50,50)

plt.figure(figsize=(8,4),dpi=150)

ax1 = plt.subplot(311)
plt.plot(x,y1,color='red')
ax2 = plt.subplot(312)
plt.plot(x,y2,color='blue')
ax3 = plt.subplot(313)
plt.plot(x,y3,color='green')
plt.subplots_adjust(hspace=0.4)

plt.figure(figsize=(8,4),dpi=150)
plt.title("Sepal-Length comparison")
plt.plot(x,y1,color='red',label='setosa')
plt.plot(x,y2,color='blue',label='versicolor')
plt.plot(x,y3,color='green',label='virginica')
plt.legend()

plt.show()
```
